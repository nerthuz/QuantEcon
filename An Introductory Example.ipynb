## Plotting a White Noise Process

```python
import numpy as np
import matplotlib.pyplot as plt

ϵ_values = np.random.randn(100)
plt.plot(ϵ_values)
plt.show()
```

## A Version with a For Loop

```python
ts_length = 100
ϵ_values = [] # empty list

for i in range(ts_length):
    e = np.random.randn()
    ϵ_values.append(e)

plt.plot(ϵ_values)
plt.show()
```

## A Version with a While Loop

```python
ts_length = 100
ϵ_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    ϵ_values.append(e)
    i = i + 1
plt.plot(ϵ_values)
plt.show()
```

## Another Application

```python
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty Numpy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
α = 0.9

X = np.empty(T+1)

for i in range(T):
    X[i+1] = α * X[i] + np.random.randn()

plt.plot(X)
plt.show()
```

## 3.2

```python
α_values = [0.1, 0.8, 0.98]

T = 200


for α in α_values:
    
    X = np.empty(T+1)
    for i in range(T):
        X[i+1] = α * X[i] + np.random.randn()
    
    plt.plot(X, label=f'$\\alpha = {α}$')

plt.legend()
plt.show()
```

## 3.3

```python
T = 200
α = 0.9

ϵ_values = np.random.randn(T)
X = np.empty(T+1)

for t in range(T-1):
    X[t+1] = α * np.abs(X[t]) + ϵ_values[t+1]

plt.plot(X)
plt.show()
```

```python
import numpy as np


T = 1000000
circle = 0

for i in range(T):
    a = np.random.uniform()
    b = np.random.uniform()
            
    hyp = np.sqrt(a**2+b**2)
    if hyp <= 1:
        circle += 1
        
pi = 4*circle/T
    
print(pi)
```
