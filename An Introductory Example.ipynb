## Plotting a White Noise Process

```python
import numpy as np
import matplotlib.pyplot as plt

_values = np.random.randn(100)
plt.plot(_values)
plt.show()
```

## A Version with a For Loop

```python
ts_length = 100
_values = [] # empty list

for i in range(ts_length):
    e = np.random.randn()
    _values.append(e)

plt.plot(_values)
plt.show()
```

## A Version with a While Loop

```python
ts_length = 100
_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    _values.append(e)
    i = i + 1
plt.plot(_values)
plt.show()
```

## Another Application

```python
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty Numpy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```
